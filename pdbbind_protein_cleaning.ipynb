```python
import pandas as pd
from pymol import cmd
```

```python
pdb_to_uniprot_csv = pd.read_csv('pdb_chain_uniprot.csv', header=0, skiprows=1)
```

```python
def pdb_to_uniprot(pdb_id, uniprot_csv):
    raw_data = uniprot_csv[uniprot_csv['PDB'] == pdb_id.lower()][['SP_PRIMARY', 'CHAIN']].values
    chain_uniprot_id_dict = {}
    for i in raw_data:
        chain_uniprot_id_dict[i[1]] = i[0]
    return chain_uniprot_id_dict
```

```python
casf_apo_structures = pd.read_csv('casf_2016_apo_ids.csv', index_col = 0)
casf_pdbs = list(casf_apo_structures[casf_apo_structures['Apo'] != 'None'].index)
casf_pdbs = [c.lower() for c in casf_pdbs]
print(len(casf_pdbs))
```

```python
def pdb_list_to_txt_file(file_name, pdb_list): 
    with open(file_name, 'w') as f:
        for p in pdb_list:
            f.write(p + '\n')
```

```python
pdb_list_to_txt_file('casf_2016_apo_only_ids.txt', casf_pdbs)
```

```python
casf_apo_structures_only = casf_apo_structures[casf_apo_structures['Apo'] != 'None']
casf_apo_structures_only_dict = {}
for pdb in casf_apo_structures_only.index:
    casf_apo_structures_only_dict[pdb.lower()] = casf_apo_structures_only[casf_apo_structures_only.index == pdb].values[0][0]
```

```python
def load_crystal_receptor(pdb_id):
    cmd.load(f'../pdbbind_2020_general/{pdb_id}/{pdb_id}_protein_cleaned.pdb')
```

```python
def load_crystal_ligand_pose(pdb_id):
    cmd.load(f'../pdbbind_2020_general/{pdb_id}/{pdb_id}_ligand_pymol.pdb')
```

```python
def fetch_apo_structure(pdb_id, dict_):
    cmd.fetch(dict_[pdb_id])
    cmd.save(f'test_apo/{pdb_id}_protein_apo.pdb', dict_[pdb_id])
    cmd.delete('all')
```

```python
def load_apo_structure(pdb_id):
    cmd.load(f'test_apo/{pdb_id}_protein_apo.pdb')
```

```python
def get_chain_ids_and_compare(pdb_id):
    load_crystal_receptor(pdb_id)
    load_apo_structure(pdb_id)
    crystal_chains = cmd.get_chains(f'{pdb_id}_protein_cleaned')
    apo_chains = cmd.get_chains(f'{pdb_id}_protein_apo')
    cmd.delete("all")
    return (crystal_chains, apo_chains)
```

```python
for c in casf_pdbs:
    fetch_apo_structure(c, casf_apo_structures_only_dict)
```
